```python
from keras.applications.vgg16 import VGG16
import matplotlib.image as mpimg
from keras import backend as K
import matplotlib.pyplot as plt

from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np
from keras.preprocessing import image

import pandas as pd
from keras.applications.vgg16 import decode_predictions

from keras.models import Sequential, Model, load_model
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization, GlobalAveragePooling2D, MaxPool2D, Lambda
from keras.applications.vgg16 import VGG16

import tensorflow.keras.layers

K.clear_session ()

model = Sequential()
model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu", name="conv2_2"))

#model.add(Lambda(custom_layer1, name="lambda_layer1"))

model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu", name="conv2_4"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu", name="conv2_7"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu", name="conv2_10"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu", name="conv2_13"))
#model.add(Lambda(custom_layer_last, name="lambda_layer_last"))

model.add(MaxPool2D(pool_size=(2,2),strides=(2,2), name="last_max_pool"))

#model.add(Lambda(custom_layer, name="lambda_layer2"))
model.add(Flatten())
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=1000, activation="softmax", name="predictions"))

print("Model created")

# model.compile(loss='categorical_crossentropy', optimizer='sgd',
#               metrics=['accuracy'], experimental_run_tf_function=False)
# print("Model compile")

VGGmodel = VGG16(weights='imagenet', include_top=True)
model.set_weights(VGGmodel.get_weights())
# model = VGG16(weights='imagenet', include_top=True)

model.summary()
```

```python

from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np

from keras import backend as K
import os
arryWeights_last_After =[]
import scipy as sp

from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np
from keras.preprocessing import image
from tensorflow.keras import models
            
import tensorflow as tf

from tensorflow.keras.preprocessing import image


for i in range(512):
  arryWeights_last_After.append(i)
# path2 = 'ClassesDataSet_OrigNames'
imgCount_init=0
imgCount = 0
hmpCount = 0
# img_path = 'Mountain_Bike/[image-net.org][27]443691662_09f12b8e37.jpg'
# folder_Images="classes_50/"
# folder_Images="FGSM_Untargeted/"
# folder_Images="PGD_Untargeted/"
# folder_Images="BIM_Untargeted/"
# folder_Images="FGSM_Targeted/"
#folder_Images="BIM_Targeted/"
folder_Images="PGD_Targeted/"

# folder_Images="Prototypes/"

imagesFileNames = []
for filename in os.listdir(folder_Images):
   for imgFileName in os.listdir(folder_Images+filename):
      imgCount_init = 0
      print('img file:', imgFileName)

      img_path = folder_Images+filename + '/'+imgFileName
      img = image.load_img (img_path, target_size=(224, 224))
      x = image.img_to_array (img)
      x = np.expand_dims (x, axis=0)
      x = preprocess_input (x)

      preds = model.predict(x)
      
      classnames = decode_predictions(preds)[0]
#       print(classnames)
      classname = classnames[0][0]
#       print("pred : ", classname)
      known_Id = classname
      element_found=0
      weights_dic_Conv2_13= {}
  #     print ('knownId:', known_Id)
      for key, value in dict_Classes.items():
          if known_Id == key : #classs_Id 
              print("element found")
              element_found=1
              weights_dic_Conv2_13.update(value)
              weights_dic_Conv2_13 = sorted (weights_dic_Conv2_13.items ())
              sorted (weights_dic_Conv2_13, key=lambda kv: kv[1], reverse=True)
              break;
      if element_found==1:
            for i in range(512):
                arryWeights_last_After[i]=0
          ###### For all classes to consider only causal features whose CI > 0
            for i in range(512):
              if weights_dic_Conv2_13[i][1][0] > 0:
                  arryWeights_last_After[i] = 1
              else:
                  arryWeights_last_After[i] = 0
            img_path = folder_Images+filename + '/'+imgFileName
            img = image.load_img (img_path, target_size=(224, 224))
            x = image.img_to_array (img)
            x = np.expand_dims (x, axis=0)
            x = preprocess_input (x)

            preds = model.predict(x)
            activatiol_layer = model.get_layer('conv2_13')

            argmax = np.argmax(preds[0])
            output = model.output[:, argmax]
            grads = K.gradients(output, activatiol_layer.output)[0]
            pooled_grads = K.mean(grads, axis=(0, 1, 2))

            
            iterate = K.function([model.input], [pooled_grads, activatiol_layer.output[0]])
            pooled_grads_value, conv_layer_output_value = iterate([x])
            
            # Similar to CAM principle, however we project here only causal features based on arryWeights_last_After
            for i in range(512):
                conv_layer_output_value[:, :, i] *= arryWeights_last_After[i]

            heatmap = np.mean(conv_layer_output_value, axis=-1)
            heatmap = np.maximum(heatmap, 0)
            heatmap /= np.max(heatmap)
            import cv2
            img = cv2.imread(img_path)
            heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
            heatmap = np.uint8(255 * heatmap)
            heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

            hif = .8
            superimposed_img = heatmap * hif + img
            # store results: superimposed attention maps on images
            directory_class_imposed = 'superimposed_img_Folder/' + filename
            print("directory_class_imposed:", directory_class_imposed)
            isExist = os.path.exists(directory_class_imposed)
            if not isExist:
               os.makedirs(directory_class_imposed)
            output =  directory_class_imposed + '/' + imgFileName
            print (output)

            cv2.imwrite(output, superimposed_img)


            directory_class_htmp = 'HeatMaps_Folder/' + filename
            isExist = os.path.exists(directory_class_htmp)
            if not isExist:
               os.makedirs(directory_class_htmp)

        #### Added for saving heatmap
            output_heatM = directory_class_htmp + '/' + imgFileName
#                 print ('dest file', 'HeatMaps_Folder/' + filename + '/' + imgFileName)
            cv2.imwrite(output_heatM, heatmap)

            import os
            import shutil

            hmpCount+=1
      imgCount+=1
      imgCount_init+=1
      print("image until now:", imgCount)
```
